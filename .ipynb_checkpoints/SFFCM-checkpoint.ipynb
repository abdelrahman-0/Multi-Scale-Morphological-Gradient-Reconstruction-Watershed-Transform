```python
import numpy as np
import imutils
import cv2
from skimage.color import rgb2lab
from math import sqrt
from random import random as r
import matplotlib.pyplot as plt
from skimage.morphology import erosion, dilation, opening, closing, disk, reconstruction
from skimage.segmentation import mark_boundaries

from scipy import ndimage as ndi
from skimage.segmentation import watershed
from skimage.feature import peak_local_max
from random import uniform
```

```python
def show(image, title='title'):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def show_inplace(image):
    plt.axis('off')
    plt.imshow(image)
    plt.show()
    
def show_inplace_g(img):
    plt.axis('off')
    plt.imshow(img, cmap='gray')
    plt.show()
    
def normalize_m(matrix):
    minimum = min([min(row) for row in matrix])
    temp = matrix - minimum
    maximum = max([max(row) for row in temp])
    temp = temp / maximum
    return temp
    
def full_grad(sobelx, sobely):
    result = np.sqrt(sobelx*sobelx + sobely*sobely)
    return normalize_m(result)
    
def mgr(img, kernel):
    e = erosion(img, kernel)
    r1 = reconstruction(e, img, method='dilation')
    d = dilation(r1, kernel)
    r2 = reconstruction(d ,r1, method='erosion')
    return r2

def mmgr(img, min_radius=1, max_iterations=50):
    result = np.zeros((img.shape)).astype(img[0][0])
    prev = result.copy()
    iterations = 0
    for i in range(max_iterations):
        prev = result
        se = disk(min_radius+i)
        gr = mgr(img, se)
        result = np.maximum(result, gr)
        if (prev == result).all():
            break
        iterations += 1
        
    print('Number of iterations till convergence: ', iterations)
    return result

def find_local_minima(img, radius=2):
    rows, cols = img.shape
    minima = []
    for y in range(radius, rows-radius):
        for x in range(radius, cols-radius):
            neighbourhood = img[y-radius:y+radius+1, x-radius:x+radius+1].flatten()
            neighbourhood_minimum = np.min(neighbourhood)
            if img[y][x] == neighbourhood_minimum:
                minima.append(np.array([x,y]))          
    return np.array(minima)

# use mean to colour the superpixel
def color_superpixels(superpixels, original):
    result = original.copy()
    rows, cols = superpixels.shape
    
    for val in np.unique(superpixels):
        pos = []
        colors = []
        for y in range(rows):
            for x in range(cols):
                if(superpixels[y][x] == val):
                    pos.append([x, y])
                    colors.append(original[y][x])
        
        avg_color = np.zeros(3)
        
        for c in colors:
            avg_color[0] += c[0]
            avg_color[1] += c[1]
            avg_color[2] += c[2]
            
        avg_color /= len(colors)
                
        for p in pos:
            result[p[1], p[0]] = avg_color
    return result
```

### Gradient image (+ gaussian filtering)

```python
# img = cv2.imread('starfish.png')
img = cv2.imread('../Deep Learning/home/abdelhaa/Desktop/Deep Learning/Dataset/Images/15011.jpg')

#OpenCV represents RGB images as multi-dimensional numpy array but in reverse order!
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows, cols, channels = img.shape

print(cols, 'x', rows)

gauss = cv2.GaussianBlur(img,(5, 5),1.0, borderType=cv2.BORDER_REPLICATE)
Lab = rgb2lab(gauss)

L, a, b = cv2.split(Lab)

L = normalize_m(L)
a = normalize_m(a)
b = normalize_m(b)

L_sobelx = cv2.Sobel(L, cv2.CV_64F, 1, 0, ksize=3)
L_sobely = cv2.Sobel(L, cv2.CV_64F, 0, 1, ksize=3)

a_sobelx = cv2.Sobel(a, cv2.CV_64F, 1, 0, ksize=3)
a_sobely = cv2.Sobel(a, cv2.CV_64F, 0, 1, ksize=3)

b_sobelx = cv2.Sobel(b, cv2.CV_64F, 1, 0, ksize=3)
b_sobely = cv2.Sobel(b, cv2.CV_64F, 0, 1, ksize=3)

L_full = np.square(full_grad(L_sobelx, L_sobely))
a_full = np.square(full_grad(a_sobelx, a_sobely))
b_full = np.square(full_grad(b_sobelx, b_sobely))

gradient = (np.sqrt(L_full + a_full + b_full))

gradient = np.clip(255 * gradient, 0 ,255).astype(np.uint8)

show_inplace(img)
```

```python
show_inplace_g(gradient)
```

### MMGR gradient

```python
morph_gradient = mmgr(gradient, min_radius = 3)
```

### Watershed

```python
# noise removal
WIDTH = 3

kernel = np.ones((WIDTH, WIDTH),np.uint8)

# denoised = cv2.morphologyEx(morph_gradient,cv2.MORPH_CLOSE, kernel, iterations = 3)
denoised = cv2.morphologyEx(morph_gradient,cv2.MORPH_OPEN, kernel, iterations = 1)
# denoised = cv2.dilate(morph_gradient, kernel, iterations = 1)
# denoised = dilation(morph_gradient, kernel)
# denoised = closing(morph_gradient, kernel)

markers = watershed(denoised)

result = color_superpixels(markers, img)

print('Number of superpixels', len(np.unique(markers)))

show_inplace(result)
markers -= 1
```

```python
contours = mark_boundaries(np.zeros((rows, cols)), markers, color=[1,1,1])
show_inplace(contours)

denoised2 = cv2.morphologyEx(gradient,cv2.MORPH_OPEN, kernel, iterations = 3)
markers2 = watershed(denoised2)

contours2 = mark_boundaries(np.zeros((rows, cols)), markers2, color=[1,1,1])
show_inplace(contours2)
```

# Phase 2

```python
def fcm_superpixels(img, markers, c=2, m=2, eta=0.00001, max_iterations = 500):
    rows, cols, channels = img.shape
    labels = np.unique(markers)
    num_regions = len(labels)
    
    # Exponent
    exponent = -2/(m-1)

    # Region size (number of pixels per region)
    region_size = np.zeros(num_regions)

    for i in labels:
        region_size[i] = np.count_nonzero(markers == i)
        
    # Region vector sum
    region_sum = np.zeros((num_regions, channels)).astype(float)
    
    for i in range(num_regions):
        region_sum[i] = np.sum(img[markers == i], axis=0)
        
    U = np.random.rand(c, num_regions).astype(float)

    for i in range(num_regions):
        U[:, i] /= np.sum(U[:, i])
    
    # Matrix of cluster centers. Each row contain 3 elements which are the coordinates of the center
    cluster_centers = np.zeros((c, channels)).astype(float)
    
    prev = U.copy()

    
    # Main LOOP begins
    for i in range(max_iterations):
        prev = U.copy()

        # Recalculate cluster centers
        for j in range(c):
            numerator = denominator = 0.0

            for k in range(num_regions):
                u_klm = U[j][k] ** m
                numerator += u_klm * region_sum[k]
                denominator += u_klm * region_size[k]

            cluster_centers[j] = numerator / denominator

        # Recalculate membership matrix
        for j in range(c):
            center = cluster_centers[j]

            for k in range(num_regions):
                region_avg = region_sum[k] / region_size[k]
                diff = region_avg - center
                numerator = np.sqrt(diff.dot(diff)) ** exponent

                denominator = 0.0
                for l in cluster_centers:
                    diff = region_avg - l
                    denominator += np.sqrt(diff.dot(diff)) ** exponent
                U[j][k] = numerator/denominator


        if(np.max(U-prev) <= eta):
            break
            
    # Assign each superpixel to the cluster which has the highest membership
    assignments = np.argmax(U, axis=0)

    clustered = np.zeros((rows, cols))

    for i in range(rows):
        for j in range(cols):
            clustered[i][j] = assignments[markers[i][j]]
    return clustered
```

```python
res = fcm_superpixels(Lab, markers, c = 2)
show_inplace(res)
```
